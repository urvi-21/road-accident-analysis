### 17. Predictive Modeling: Casualty Severity Prediction 

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
# Prepare data
features = data[['age_of_casualty', 'sex_of_casualty', 'vehicle_reference']]
target = data['casualty_severity']
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)
```

```python
# Train model
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Predict and evaluate
predictions = model.predict(X_test)
print(classification_report(y_test, predictions))
```

### 18. Clustering Casualties by Age and Severity (K-Means)

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Prepare data
X = data[['age_of_casualty', 'casualty_severity']].dropna()
X_scaled = StandardScaler().fit_transform(X)
```

```python
# K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=0)
clusters = kmeans.fit_predict(X_scaled)

plt.figure(figsize=(10, 6))
plt.scatter(X['age_of_casualty'], X['casualty_severity'], c=clusters, cmap='viridis', alpha=0.6)
plt.xlabel('Age of Casualty')
plt.ylabel('Casualty Severity')
plt.title('Clustering of Casualties by Age and Severity')
plt.colorbar()
plt.show()
```

### 21. Cluster Analysis: Accident Patterns

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Feature scaling
features = data[['age_of_casualty', 'casualty_severity']]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
# Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(scaled_features)
data['cluster'] = clusters
```

```python
# Plot clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age_of_casualty', y='casualty_severity', hue='cluster', data=data, palette='viridis')
plt.title('Clusters of Accident Patterns')
plt.show()
```
